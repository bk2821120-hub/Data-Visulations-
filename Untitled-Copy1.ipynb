```python
Display_According_To_Size_Sales=df.groupby("Outlet Size")["Sales"].sum().reset_index()
Display_According_To_Size_Sales
```

```python
Bar_Display_According_To_Size_Sales=px.bar(Display_According_To_Size_Sales,x="Outlet Size",y="Sales",title="Sales According To Size",color="Outlet Size")
Bar_Display_According_To_Size_Sales
```

```python

































Bar_Display_According_To_Size_Sales=px.pie(Display_According_To_Size_Sales,values="Sales",names="Outlet Size", hole=0.5,title="Sales According To Size",color="Outlet Size")
Bar_Display_According_To_Size_Sales
```

```python
Bar_Display_According_To_Size_Sales=px.scatter(data_frame=Display_According_To_Size_Sales,x="Outlet Size",y="Sales",trendline="ols",title="Sales According To Size",color="Outlet Size")
Bar_Display_According_To_Size_Sales
```
